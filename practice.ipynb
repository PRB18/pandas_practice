```python
def get_era(Year):
    if Year<1980:
        return "Classic"
    elif Year<2010:
        return "Modern"
    else:
        return "Contemporary"
bollywood["Era"] = bollywood["Year"].apply(get_era)
bollywood[["Year","Title","Era"]].head()
```

```python
#bollywood.drop(columns="Cast",inplace=True)
bollywood.rename(columns={"Release Month":"month"},inplace=True)
```

```python
#Find all drama movies released in December after 1990, sorted by Year newest first. 
# Show only Title, Year, Director.
bollywood[(bollywood["month"] == "DEC") & (bollywood["Year"] > 1990) & (bollywood["Genre"] == "drama")][["Title", "Year", "Director"]].sort_values("Year", ascending=False)
```
