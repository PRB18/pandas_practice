```python
import pandas as pd

df = pd.read_csv("archive/bollywood.csv", encoding="latin1")
actors = pd.read_csv("archive/bollywood-actors.csv", encoding="latin1")
actress = pd.read_csv("archive/bollywood-actress.csv", encoding="latin1")
```

```python
#Add a "Role" column to actors ("Actor") and actress ("Actress"). 
# Concat both into df_talent. How many people total? Check df_talent.head() — 
# what happened to "Debut aslead role" for actor rows?
actors["Role"] = "actor"
actress["Role"] = "actress"
df_talent = pd.concat([actors,actress])
df_talent["Role"].value_counts()
df_talent.head()
```

```python
df_talent.query('`Height(in cm)` > 180').sort_values("Height(in cm)", ascending=False)
df_talent.query('Height(in cm)>180').sort_values("Height(in cm)",ascending=False)
```
